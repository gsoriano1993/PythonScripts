```python
for z in range(0,len(lista_dias)):
    dfTest=df.loc[df.periodo==int(PERIODO)].loc[df.canal==CANAL, ["hora_ingreso", 'promedio_of_mes']].reset_index(drop=True).copy()
    dfTest=dfTest.sort_values(by="hora_ingreso").reset_index(drop=True)
    dfFinal=pd.merge(dfAux2, dfTest, left_on="Hora", right_on="hora_ingreso", how="inner").drop(columns="hora_ingreso").copy()
    lista=dfAux3[dfAux3.Hora=='Pago'].stack().reset_index()[0].to_list()
    lista.append(0)
    dfFinal.loc[len(dfFinal)] = lista
    dfFinal=dfFinal.set_index('Hora')

    a = dfFinal.drop(index=["Pago"], columns=[lista_dias[z]]).values
    # number of shifts
    n = a.shape[1]
    # number of time windows
    T = a.shape[0]
    # number of workers required per time window
    d = dfFinal[lista_dias[z]].values
    # wage rate per shift
    w = dfFinal.loc["Pago", :].values.astype(int)
    # Decision variables
    y = LpVariable.dicts("num_workers", list(range(n)), lowBound=0, cat="Integer")
    # Create problem
    prob = LpProblem("scheduling_workers", LpMinimize)
    prob += lpSum([w[j] * y[j] for j in range(n)])
    for t in range(T):
        prob += lpSum([a[t, j] * y[j] for j in range(n)]) >= d[t]
    prob.solve()
    print("Status:", LpStatus[prob.status])
    df_resultados[lista_dias[z]]=0
    for i in range(0, len(df_resultados)):
        df_resultados[lista_dias[z]][i]=y[i].value()
```
